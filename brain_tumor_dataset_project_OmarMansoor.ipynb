### Packages

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator
```

### Import dataframe

```python
df = pd.read_csv('brain_tumor_dataset.csv')
```

### Boxplots w/ Statistical Annotations

```python
figure = plt.figure(figsize=(7.2,6.2))
sns.set_context('paper')
sns.set_style('white')
mpl.rcParams['font.family'] = ['sans-serif']
mpl.rcParams['font.sans-serif'] = ['Arial']

custom_palette={
    'Male': 'gray',
    'Female': '#d5322c'
}

ax = sns.boxplot(df, x='Tumor Type', y='Size (cm)', hue='Gender', gap=0.145, palette=custom_palette, legend=True, showmeans=True)
ax = sns.swarmplot(df, x='Tumor Type', y='Size (cm)', hue='Gender', dodge=True, color='black', legend=False, alpha=0.4, linewidth=0.15, edgecolor='black')
ax.legend(frameon=False, fontsize=10, title='', bbox_to_anchor=(1,0.5))
ax.spines[['right', 'top']].set_visible(False)
ax.set_title('', fontsize=12, weight='bold', loc='left')
ax.tick_params(axis='x', bottom=True, labelsize=10)
ax.tick_params(axis='y', left=True, labelsize=8)
ax.set_xlabel('')
ax.set_ylabel('Size (cm)', fontsize=10)

pairs = [(('Oligodendroglioma', 'Male'), ('Oligodendroglioma', 'Female')), (('Ependymoma', 'Male'), ('Ependymoma', 'Female')), (('Meningioma', 'Male'), ('Meningioma', 'Female')), (('Astrocytoma', 'Male'), ('Astrocytoma', 'Female')), (('Glioblastoma', 'Male'), ('Glioblastoma', 'Female'))]

annotator = Annotator(ax, pairs, data=df, x='Tumor Type', y='Size (cm)', hue='Gender')
annotator.configure(test='t-test_ind', text_format='star', loc='inside')
annotator.apply_and_annotate()
```

#### Specifying Dataframes for each Tumor Type

```python
df_Oligo = df[df['Tumor Type'] == 'Oligodendroglioma']
df_Epen = df[df['Tumor Type'] == 'Ependymoma']
df_Menin = df[df['Tumor Type'] == 'Meningioma']
df_Astro = df[df['Tumor Type'] == 'Astrocytoma']
df_Glio = df[df['Tumor Type'] == 'Glioblastoma']
```

### Grouping the Sizes into Small, Medium and Large to Conserve Space in the Legend

```python
df_Oligo['Size (cm)'] = pd.to_numeric(df_Oligo['Size (cm)'], errors='coerce')

def group_size(x):
    if x < 3:
        return "Small"
    elif x < 5:
        return "Medium"
    else:
        return "Large"
df_Oligo['Size (cm)'] = df_Oligo['Size (cm)'].apply(group_size)
```
